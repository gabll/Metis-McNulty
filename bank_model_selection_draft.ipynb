```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import pymysql
import numpy as np
import seaborn as sb
import model_selection_utils
import credentials
```

<a id='1'></a>
## Load data

```python
db = pymysql.connect(host=credentials.ip,
                     user=credentials.user,
                     passwd=credentials.password,
                     db='bank')

df = pd.read_sql("select * from bank", db)
db.close()
df.head(10)
```

<a id='2'></a>
##Data preparation

```python
#Binaryze outcome var
df['y'][df['y'] == 'yes'] = 1
df['y'][df['y'] == 'no'] = 0

features_list = list(df.columns)
features_list.remove('y')
features_list.remove('duration')  #we want to target new clients
features_list.remove('campaign')
features_list.remove('pdays')
features_list.remove('previous')
features_list.remove('poutcome')
features_list.remove('day')       #we dind't found a useful way to interpret it by now

outcome_var = 'y'
features_list.append(outcome_var)
    
df_conv = model_selection_utils.convert_features(df[features_list], outcome_var, dummies=True,
                                                 scaling=True, only_important_features=True, null_value='unknown')

X = df_conv.drop([outcome_var], 1)
Y = df_conv[outcome_var]

print 'All features: ', len(X.columns)
print 'N.rows: ', len(X)
print 'Feature list: ', X.columns
```

<a id='3'></a>
##Feature selection

```python
from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import f_classif
from sklearn.cross_validation import train_test_split

X_new = SelectKBest(f_classif, k=20).fit(X, Y).get_support()

selected_features = list(X.columns[X_new])
print "First 20 \"important\" features:\n", selected_features
```

```python
#Drop features
X = X.drop(['default_yn_no', ], 1) #it doesn't bring a lot of information (we have balance)
X = X.drop(['contact_telephone'], 1) #it's not so relevant

#Split test and training set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=42)
```

<a id='4'></a>
##k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_graph = []
y_graph = []
for k in range(1,15):
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train,Y_train)
    x_graph.append(k)
    y_graph.append(accuracy_score(Y_test, knn.predict(X_test)))

plt.xlabel('k')
plt.ylabel('Accurancy')
plt.title('k-NN accuracy (' + str(len(X_train)) + ' obs)')
plt.plot(x_graph,y_graph, label='K-NN')
plt.show()
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.cross_validation import train_test_split

X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=42)

knn = KNeighborsClassifier(n_neighbors=6).fit(X_train, Y_train)

model_selection_utils.print_model_evaluation(knn, Y_test, X_test, Y, X, cross_val_samples=2)
```

<a id='5'></a>
##Logistic regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression().fit(X_train, Y_train)

print model_selection_utils.print_cross_val_score(log_reg, X, Y, cv=2)
print model_selection_utils.print_cross_val_score(log_reg, X, Y, cv=2, scoring='precision')
print model_selection_utils.print_cross_val_score(log_reg, X, Y, cv=2, scoring='recall')
print model_selection_utils.print_cross_val_score(log_reg, X, Y, cv=2, scoring='f1')
```

<a id='6'></a>
##Decision Tree/Random Forest

```python
from sklearn import tree
import pydot
import pyparsing

dt = tree.DecisionTreeClassifier()
dt = dt.fit(X_train,Y_train)

model_selection_utils.print_model_evaluation(dt, Y_test, X_test, Y, X, cross_val_samples=2)
```

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier().fit(X_train,Y_train)
model_selection_utils.print_model_evaluation(rf, Y_test, X_test, Y, X, cross_val_samples=2)
```

```python
from sklearn.svm import SVC

svc = SVC().fit(X_train,Y_train)
print model_selection_utils.print_cross_val_score(svc, X, Y, cv=2)
print model_selection_utils.print_cross_val_score(svc, X, Y, cv=2, scoring='precision')
print model_selection_utils.print_cross_val_score(svc, X, Y, cv=2, scoring='recall')
print model_selection_utils.print_cross_val_score(svc, X, Y, cv=2, scoring='f1')
```
