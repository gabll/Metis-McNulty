```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import pymysql
import numpy as np
import seaborn as sb
import model_selection_utils
import credentials
```

<a id='1'></a>
## Load data

```python
# Source: [Moro et al., 2014] S. Moro, P. Cortez and P. Rita.
# A Data-Driven Approach to Predict the Success of Bank Telemarketing.
# Decision Support Systems, Elsevier, 62:22-31, June 2014

db = pymysql.connect(host=credentials.ip,
                     user=credentials.user,
                     passwd=credentials.password,
                     db='bank')

df = pd.read_sql("select * from bank_additional_2", db)
db.close()
df.head(10)
```

```python
df.dtypes
```

<a id='2'></a>
##Data preparation

```python
#Binaryze outcome var
df['y'][df['y'] == 'yes'] = 1
df['y'][df['y'] == 'no'] = 0

features_list = list(df.columns)
features_list.remove('y')
features_list.remove('duration') #See dataset description

outcome_var = 'y'
features_list.append(outcome_var)
    
df_conv = model_selection_utils.convert_features(df[features_list], outcome_var, dummies=True,
                                                 scaling=True, only_important_features=True, null_value='unknown')

X = df_conv.drop([outcome_var], 1)
Y = df_conv[outcome_var]
features_list_dummies = X.columns

print 'All features: ', len(X.columns)
print 'N.rows: ', len(X)
print 'Feature list: ', X.columns
```

```python
from sklearn.cross_validation import train_test_split

#Shuffle X and Y (necessary for plotting the learning curve)
X, _, Y, _ = train_test_split(X,Y, test_size=0, random_state=1)

#Split test and training set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=42)
```

<a id='3'></a>
##Baseline predictor

```python
#Baseline predictor: return yes if poutcome_success = true
X_test_w_col = pd.DataFrame(X_test, columns=list(features_list_dummies))
def baseline_1(X):
    return [1 if i == 1 else 0 for i in list(X.poutcome_success)]

print 'Accuracy baseline 1:\t%.4f' % accuracy_score(Y_test, baseline_1(X_test_w_col))
print 'Precision baseline 1:\t%.4f' % precision_score(Y_test, baseline_1(X_test_w_col))
print 'Recall baseline 1:\t%.4f' % recall_score(Y_test, baseline_1(X_test_w_col))
print 'F1 baseline 1:\t\t%.4f' % f1_score(Y_test, baseline_1(X_test_w_col))

print '\nConfusion matrix:\n', model_selection_utils.format_confusion_matrix(Y_test, baseline_1(X_test_w_col))
```

This baseline predictor has good performances: it means that only the poutcome feature brings a lot of information alone. However, if we have new clients or if we want to capture conversions from previous 'no' to new 'yes' in the outcome variable we have to expand the feature set.

<a id='4'></a>
##k-NN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

x_graph = []
y_graph = []
for k in range(1,25):
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train,Y_train)
    x_graph.append(k)
    y_graph.append(accuracy_score(Y_test, knn.predict(X_test)))

plt.xlabel('k')
plt.ylabel('Accurancy')
plt.title('k-NN accuracy (' + str(len(X_train)) + ' obs)')
plt.plot(x_graph,y_graph, label='K-NN')
plt.show()
```

```python
knn = KNeighborsClassifier(n_neighbors=17).fit(X_train, Y_train)

model_selection_utils.print_model_evaluation(knn, Y_test, X_test, Y, X, cross_val_samples=1)
```

##Logistic Regression

```python
from sklearn.linear_model import LogisticRegression

log_reg = LogisticRegression().fit(X_train, Y_train)

model_selection_utils.print_model_evaluation(log_reg, Y_test, X_test, Y, X, cross_val_samples=1)
```

##Decision Tree/Random Forest

<a id='7'></a>
##Support Vector Machine

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score


svc = SVC().fit(X_train,Y_train)

print 'Accuracy:\t%.4f' % accuracy_score(Y_test, svc.predict(X_test))
print 'Precision:\t%.4f' % precision_score(Y_test, svc.predict(X_test))
print 'Recall:\t\t%.4f' % recall_score(Y_test, svc.predict(X_test))
print 'F1:\t\t%.4f' % f1_score(Y_test, svc.predict(X_test))

plt.figure(figsize=(18,5))
plt.subplot(1, 2, 1)
model_selection_utils.plot_learning_curve(svc,'Learning curve', X, Y)
plt.subplot(1, 2, 2)
model_selection_utils.plot_roc_svc(svc,'ROC', X_test, Y_test)
plt.show()
print '\nConfusion Matrix'
print model_selection_utils.format_confusion_matrix(Y_test, svc.predict(X_test))
```

```python
#Fit again the model without features scaling

#Binaryze outcome var
df['y'][df['y'] == 'yes'] = 1
df['y'][df['y'] == 'no'] = 0

features_list = list(df.columns)
features_list.remove('y')
features_list.remove('duration') #See dataset description

outcome_var = 'y'
features_list.append(outcome_var)
    
df_conv = model_selection_utils.convert_features(df[features_list], outcome_var, dummies=True,
                                                 scaling=False, only_important_features=True, null_value='unknown')

X = df_conv.drop([outcome_var], 1)
Y = df_conv[outcome_var]
features_list_dummies = X.columns

#Shuffle X and Y (necessary for plotting the learning curve)
X, _, Y, _ = train_test_split(X,Y, test_size=0, random_state=1)

#Split test and training set
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size=.20, random_state=42)

log_reg = LogisticRegression().fit(X_train, Y_train)

coefs = dict(zip(list(features_list_dummies), list(log_reg.coef_[0])))
intercept = log_reg.intercept_
print '\n', str(list(log_reg.coef_[0]))
print log_reg.intercept_
#print ['point.' + i for i in list(features_list_dummies)] #for copy-paste to html page
```

```python
import names
import random
#reassign column names
new_df = pd.DataFrame(X, columns=list(features_list_dummies))
Y = pd.Series(Y)
#new campaign = Y became poutcome
new_df['poutcome_success'] = Y
new_df['poutcome_nonexistent'] = 0
new_df['poutcome_failure'] = Y.replace(0,1)-Y
#new campaign = campaign become previous
new_df['previous'] = new_df['campaign']
#new campaign: no contacts before during this campaign
new_df['campaign'] = 0
#new campaign has occurred one week after the last campaign
new_df['pdays'] = 5
#new campaign: initial values of the webapp
new_df[[u'month_apr', u'month_aug', u'month_dec', u'month_jul', u'month_jun',
       u'month_mar', u'month_may', u'month_nov', u'month_oct', u'month_sep']] = 0
new_df[[u'day_of_week_fri', u'day_of_week_thu', u'day_of_week_tue', u'day_of_week_wed']] = 0
new_df[u'day_of_week_mon'] = 1
new_df[u'emp_var_rate'] = -2
new_df[u'cons_price_idx'] = 93
new_df[u'cons_conf_idx'] = -25
new_df[u'euribor3m'] = 0.2
new_df[u'nr_employed'] = 4900

#new campaign: initial prediction (will change in the DSS basing on the input)
new_df['test_prediction'] = log_reg.predict_proba(new_df)[:,1]
```

```python
#descriptive columns for tips in the visualization
df_mix, _= train_test_split(df[features_list], test_size=0, random_state=1)
df_mix = pd.DataFrame(df_mix, columns=list(features_list))
new_df['job'] = df_mix['job']
new_df['marital'] = df_mix['marital']
new_df['education'] = df_mix['education']
new_df['default_yn'] = df_mix['default_yn']
new_df['housing'] = df_mix['housing']
new_df['loan'] = df_mix['loan']
new_df['contact'] = df_mix['contact']
```

```python
#sort and filter first 1500 rows
new_df = new_df.sort(['test_prediction'], ascending=False)
new_df = new_df.head(1500)
```

```python
#randomize a name and a customer code for each row
new_df['client_name'] = [names.get_full_name() for _ in range(0,1500)]
new_df['account_code'] = [random.randint(100000, 999999) for _ in range(0,1500)]
```
