```python
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import pymysql
import numpy as np
import credentials
```

```python
db = pymysql.connect(host=credentials.ip,
                     user=credentials.user,
                     passwd=credentials.password,
                     db='bank')

df = pd.read_sql("select * from bank", db)
print df.head(10)
```

```python
df.dtypes
```

```python
df.describe()
```

```python
plt.figure(figsize=(18,5))
plt.subplot(1, 3, 1)
plt.xlabel('age')
plt.boxplot(df.age, showmeans=True, showfliers=False)
plt.subplot(1, 3, 2)
plt.xlabel('balance')
plt.boxplot(df.balance, showmeans=True, showfliers=False)
plt.subplot(1, 3, 3)
plt.xlabel('duration')
plt.boxplot(df.duration, showmeans=True, showfliers=False)
plt.show()
```

```python
for i in df.columns:
    if df[i].dtypes == np.object:
        print df[[i]].groupby(i).size().order(ascending=False)
        print '\n'
```
